# Food Delivery Hackathon Analysis

This notebook analyzes food delivery data by combining multiple datasets and extracting business insights.

## Datasets Used
- orders.csv (Transactional Data)
- users.json (User Master Data)
- restaurants.sql (Restaurant Master Data)


## Data Merging Logic
- orders LEFT JOIN users on user_id
- orders LEFT JOIN restaurants on restaurant_id
- LEFT JOIN is used to retain all order records


## Analysis Performed
- Revenue analysis by city and membership type
- Cuisine-wise average order value
- User behavior analysis
- Seasonal revenue trends

## Key Insights
- Hyderabad generated the highest revenue from Gold members
- Italian cuisine has the highest average order value
- Gold members contributed approximately 50% of total orders
